```python
batch = 10
ind = 3
inh = 180
inw = 390
inc = 3
upscale = 2
```

```python
mysr = MySR(upscale)
mysr.build(input_shape=(batch,ind,inh,inw,inc))
print(mysr.summary())

mysr.load_weights("./model/Mysr2x")
```

```python
temp_dir = "./test_sr/"
if bool(1-os.path.exists(temp_dir)):
    os.mkdir(temp_dir)
    print("temp_dir not exists, create one.")

lr_path = "../data/test"
lr_list = os.listdir(lr_path)
lr_list.sort(key=lambda x:int(x[5:-4]))

for i in range(len(lr_list)):
    inputs = []
    try:
        path = os.path.join(lr_path,lr_list[i-2])
        img = Image.open(path)
        img_arr = np.array(img)
    except:
        path = os.path.join(lr_path,lr_list[i])
        img = Image.open(path)
        img_arr = np.array(img)
        img_arr[...] = 0
    inputs.append(img_arr)
    
    try:
        path = os.path.join(lr_path,lr_list[i-1])
        img = Image.open(path)
        img_arr = np.array(img)
    except:
        path = os.path.join(lr_path,lr_list[i])
        img = Image.open(path)
        img_arr = np.array(img)
        img_arr[...] = 0
    inputs.append(img_arr)
    
    path = os.path.join(lr_path,lr_list[i])
    img = Image.open(path)
    img_arr = np.array(img)
    inputs.append(img_arr)
    inputs = np.array(inputs)
    
    inputs = np.float32(inputs[np.newaxis,...])
    pred = mysr.predict(inputs)[0]
    pred = Image.fromarray(np.uint8(pred),"RGB")
    pred.save(os.path.join(temp_dir,lr_list[i]))
    print("progress %s" % lr_list[i])
```
